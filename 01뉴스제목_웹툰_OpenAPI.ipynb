### 1. 뉴스 제목 추출하기

```python
# IT/과학 뉴스 
req_param = {
    'sid': 105
}
# 
url = 'https://news.naver.com/section/{sid}'.format(**req_param)
print(url)

url2 = 'https://news.naver.com/section/105'

# 요청 헤더 설정 : 브라우저 정보
req_header = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/134.0.0.0 Safari/537.36'}

# requests 의 get() 함수 호출하기 
res = requests.get(url, headers=req_header)
print(res)
print(type(res))
print(res.status_code, res.ok)

# 응답(response)이 OK 이면
if res.ok:
 # 응답 (response)에서 text 추출 -source 보기
 html = res.text
 #print(html)
 # BeautifulSoup 객체 생성  
 soup = BeautifulSoup(html, 'html.parser')
 #print(soup)
 # CSS 선택자
 #div class='sa_text'>
 print(len(soup.select("div.sa_text a[href*='mnews/article']")))
 # print(soup.select("div.sa_text a[href*='mnews/article']"))
 a_tag_list = soup.select("div.sa_text a[href*='mnews/article']")
 print(type(a_tag_list)) # ResultSet
 print(a_tag_list) #[<a href="http://>뉴스제목</a>] ResultSet 객체, <a></a> : Tag 객체체
 for a_tag in a_tag_list:
     #print(type(a_tag))
     title = a_tag.text.strip()
     # href 속성의 값을 가져오기기
     link = a_tag['href']
     print(link, title)

 # <a> 태그 리스트 순회하기    
else:
 # 응답(response)이 Error 이면 status code 출력    
 print(f'에러코드 = {res.status_code}')
```

### 1.1 뉴스제목 추출하는 함수 선언하기

```python
import requests
from bs4 import BeautifulSoup

section_dict = {
    100: "정치",
    101: "경제",
    102: "사회",
    103: "생활/문화",
    104: "세계",
    105: "IT/과학",
}


def print_news(sid):  # print_new(103,'생활/문화')
    section = section_dict[sid]
    req_param = {"sid": sid}
    #
    url = f"https://news.naver.com/section/{sid}"
    print(f"=====> {url} {section} 뉴스 <=====")

    # 요청 헤더 설정 : 브라우저 정보
    req_header = {
        "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/134.0.0.0 Safari/537.36"
    }

    # requests 의 get() 함수 호출하기
    res = requests.get(url, headers=req_header)
    print(res)
    print(type(res))
    print(res.status_code, res.ok)


 # 응답(response)이 OK 이면
    if res.ok:
     # 응답 (response)에서 text 추출 -source 보기
     html = res.text
     # print(html)
     # BeautifulSoup 객체 생성
     soup = BeautifulSoup(html, "html.parser")
     # print(soup)
     # CSS 선택자
     # div class='sa_text'>
     print(len(soup.select("div.sa_text a[href*='mnews/article']")))
     # print(soup.select("div.sa_text a[href*='mnews/article']"))
     a_tag_list = soup.select("div.sa_text a[href*='mnews/article']")
     print(type(a_tag_list))  # ResultSet
     print(a_tag_list) # [<a href="http://>뉴스제목</a>] ResultSet 객체, <a></a> : Tag 객체체
     for a_tag in a_tag_list:
         # print(type(a_tag))
         title = a_tag.text.strip()
         # href 속성의 값을 가져오기기
         link = a_tag["href"]
         print(link, title)

     # <a> 태그 리스트 순회하기
    else:
     # 응답(response)이 Error 이면 status code 출력
     print(f"에러코드 = {res.status_code}")
         
         
print_news(105)
```
