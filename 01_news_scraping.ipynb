```python
#링크와 제목을 출력
import requests
from bs4 import BeautifulSoup

url = "https://news.daum.net/economy"
req_header = {
    "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/134.0.0.0 Safari/537.36"
}
res = requests.get(url, headers=req_header)
print(res)
print(type(res))
print(res.status_code, res.ok)
res.encoding = "utf-8"


if res.ok:
    html = res.text
    soup = BeautifulSoup(html, "html.parser")
    print(len(soup.select("a[href*='v/']")))
    a_tag_list = soup.select("a[href*='v/']")
    print(type(a_tag_list))
    print(a_tag_list)
    for a_tag in a_tag_list:
        title = a_tag.text.strip()
        link = a_tag["href"]
        print(link, title)
else:
    print(f"에러코드 = {res.status_code}")
```

```python
#함수로 작성하기
import requests
from bs4 import BeautifulSoup


section_dict = {
    '기후/환경': 'climate',
    '사회': 'society',
    '경제': 'economy',
    '정치': 'politics',
    '국제': 'world',
    '문화': 'culture',
    '생활': 'life',
    'IT/과학': 'tech',
    '인물': 'people',
}


def print_news(section_name):
    sid = section_dict.get(section_name)
    url = f"https://news.daum.net/{sid}"
    
    print(f"======> {url} {section_name} 뉴스 <======")
    
    req_header = {
        "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/134.0.0.0 Safari/537.36"
    }
    res = requests.get(url, headers=req_header)
    print(res)
    print(type(res))
    print(res.status_code, res.ok)
    res.encoding = "utf-8"
    
    if res.ok:
     html = res.text
     soup = BeautifulSoup(html, "html.parser")
     print(len(soup.select("a[href*='v/']")))
     a_tag_list = soup.select("a[href*='v/']")
     print(type(a_tag_list))
     print(a_tag_list)
     for a_tag in a_tag_list:
         title = a_tag.text.strip()
         link = a_tag["href"]
         print(link, title)
    else:
     print(f"에러코드 = {res.status_code}")
   

print_news('경제')
print_news('사회')
```
