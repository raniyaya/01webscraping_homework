### Nate 뉴스기사 제목 스크래핑하기

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
from IPython.display import Image, display

url = 'https://m.news.nate.com/section?mid=m03&sq=1088503'  # 모바일 뉴스 주소
req_header = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/134.0.0.0 Safari/537.36'
}
res = requests.get(url, headers=req_header)
if res.ok:
    soup = BeautifulSoup(res.text, 'html.parser')
    li_tag_list = soup.select('ul.mlt01 li')


    for li_tag in li_tag_list:
        a_tag = li_tag.find('a')
        if not a_tag:
            continue

        title = a_tag.text.strip()
        link = urljoin(url, a_tag['href'])

        print(link)
        print(title)

        # 이미지 태그 추출
        img_tag = li_tag.find('img')
        if img_tag:
            img_src = urljoin(url, img_tag['src'])
            print("이미지 링크:", img_src)
            display(Image(img_src))
        else:
            print("이미지 없음")
        print()
else:
    print(f'에러코드: {res.status_code}')
```
